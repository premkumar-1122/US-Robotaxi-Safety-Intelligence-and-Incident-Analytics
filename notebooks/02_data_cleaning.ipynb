## Feature Engineering

```python
df_clean['Incident Year'] = (
    df_clean['Incident Date'].dt.year
)

df_clean['Incident Month'] = (
    df_clean['Incident Date'].dt.month
)
```

```python
df_clean['Incident Month Name'] = (
    df_clean['Incident Date'].dt.month_name()
)

df_clean['Incident Quarter'] = (
    df_clean['Incident Date'].dt.quarter
)
```

```python
df_clean['Reporting Delay Months'] = (
    (
        df_clean['Report Submission Date'].dt.year
        - df_clean['Incident Date'].dt.year
    ) * 12
    +
    (
        df_clean['Report Submission Date'].dt.month
        - df_clean['Incident Date'].dt.month
    )
)
```

### Cleaning weather columns

```python
weather_cols = [
    'Weather - Clear',
    'Weather - Rain',
    'Weather - Fog/Smoke/Haze'
]

df_clean[weather_cols].eq('Y').sum(axis=1).value_counts()
```

```python
def consolidate_weather(row):

    if row['Weather - Clear'] == 'Y':
        return 'Clear'

    elif row['Weather - Rain'] == 'Y':
        return 'Rain'

    elif row['Weather - Fog/Smoke/Haze'] == 'Y':
        return 'Fog/Smoke/Haze'

    else:
        return 'Unknown'
```

```python
df_clean['Weather Condition'] = (
    df_clean.apply(consolidate_weather, axis=1)
)
```

```python
df_clean['Weather Condition'].value_counts()
```

```python
df_clean = df_clean.drop(columns=weather_cols)
```

```python
df_analysis = df_clean.copy()
```

## Removing Duplicates

### Severity Normalization

```python
df_analysis['Highest Injury Severity Alleged'].value_counts()
```

```python
severity_map = {
    'Property Damage. No Injured Reported': 'No Injury',
    'No Injured Reported': 'No Injury',

    'Minor W/O Hospitalization': 'Minor Injury',
    'Minor W/ Hospitalization': 'Minor Injury',

    'Moderate W/O Hospitalization': 'Moderate Injury',
    'Moderate W/ Hospitalization': 'Moderate Injury',

    'Fatality': 'Fatality',

    'Unknown': 'Unknown'
}

df_analysis['Severity Group'] = (
    df['Highest Injury Severity Alleged']
    .map(severity_map)
)

df_analysis['Severity Group'].value_counts()
```

### Counterpart Normalization

```python
df_analysis['Crash With'].value_counts()
```

```python
counterpart_map = {

    'Passenger Car': 'Passenger Vehicle',
    'SUV': 'Passenger Vehicle',
    'Pickup Truck': 'Passenger Vehicle',
    'Van': 'Passenger Vehicle',

    'Heavy Truck': 'Commercial Vehicle',
    'Bus': 'Commercial Vehicle',

    'Motorcycle': 'Two-Wheeler',

    'Non-Motorist: Cyclist': 'Vulnerable Road User',
    'Non-Motorist: Pedestrian': 'Vulnerable Road User',
    'Non-Motorist: Scooter - Skateboard': 'Vulnerable Road User',
    'Non-Motorist: Other': 'Vulnerable Road User',

    'Pole / Tree': 'Fixed Infrastructure',
    'Other Fixed Object': 'Fixed Infrastructure',

    'Animal': 'Animal',

    'First Responder Vehicle': 'Emergency Vehicle',

    'Other, see Narrative': 'Other / Unknown'
}

df_analysis['Crash Counterpart Group'] = (
    df['Crash With']
    .map(counterpart_map)
)
```
